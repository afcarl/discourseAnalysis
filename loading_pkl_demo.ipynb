```python
from os import listdir
from os.path import join as joinpath
```

```python
path0 = 'data/narrative/'
path1 = 'data/argumentative/'
path2 = 'data/informative/'
doclist0 = [joinpath(path0, fname) for fname in listdir(path0) if fname.endswith('.xml')]
doclist1 = [joinpath(path1, fname) for fname in listdir(path1) if fname.endswith('.xml')]
doclist2 = [joinpath(path2, fname) for fname in listdir(path2) if fname.endswith('.xml')]
```

```python
def get_docs(doclist):
    docs = []
    for f in doclist:
        doc = open(f, 'r')
        content = doc.read()
        docs.append((content,f))
    return docs
```

```python
docs0 = get_docs(doclist0)
docs1 = get_docs(doclist1)
docs2 = get_docs(doclist2)
```

```python
import BeautifulSoup as BeautifulSoup
import copy
import numpy as np

def build_count_tags_vect(doc):
    soup = BeautifulSoup.BeautifulSoup(doc)
    tags = soup.findAll('pos')
    dico = {}
    for t in tags:
        if t.text not in dico.keys():           
            dico[t.text] = 1
        else:
            dico[t.text] += 1
    return dico

def build_norm_tags_vects(count_vect):
    c_ve = copy.deepcopy(count_vect)
    print 'deepcopied'
    for i in range(len(c_ve)):
        total = np.sum(c_ve[i][0].values())
        for j in c_ve[i][0].keys():
            c_ve[i][0][j] = float(c_ve[i][0][j])/total        
    return c_ve

def build_count_tags_vects(docs):
    return [(build_count_tags_vect(doc[0]),doc[1]) for doc in docs]
```

```python
narr = build_count_tags_vects(docs0)
arg = build_count_tags_vects(docs1)
inf = build_count_tags_vects(docs2)
```

```python
y_nar = [0 for t in narr]
y_arg = [1 for t in arg]
y_inf = [2 for t in inf]
```

```python
docs = narr + arg + inf
y = y_nar + y_arg + y_inf
import cPickle as pickle
pickle.dump(docs,open('pos_count_vects.pkl','w'))
pickle.dump(y,open('y_pos_count_vects.pkl','w'))
```

```python
norm = build_norm_tags_vects(docs)
pickle.dump(norm,open('pos_norm_vects.pkl','w'))
```
